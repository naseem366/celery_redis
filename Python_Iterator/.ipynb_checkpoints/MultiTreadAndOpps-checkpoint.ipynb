### What is multi-threading? 

#### Multithreading is a programming technique that allows concurrent execution of multiple threads within a single process.

##### Threads are lightweight processes that share the same memory space.
##### Python's 'threading' module provides a way to create and manage threads.
##### Example: Using threading to perform I/O operations while keeping the main program responsive.
##### Multithreading can improve performance in I/O-bound tasks but may not significantly speed up CPU-bound tasks due to the Global Interpreter Lock (GIL).
##### Example: A web server handling multiple requests simultaneously using threads.



   

What are the OOPS concepts

Object-oriented programming concepts that focus on classes and objects.

    Encapsulation: Bundling data and methods that operate on the data into a single unit (class).

    Inheritance: Ability of a class to inherit properties and behavior from another class.

    Polymorphism: Ability to present the same interface for different data types.

    Abstraction: Hiding the complex implementation details and showing only the necessary features.



| Pillar        | Purpose                                                   |
| ------------- | --------------------------------------------------------- |
| Encapsulation | Bundle data + methods; hide internal details              |
| Abstraction   | Hide complex details; expose only necessary parts         |
| Inheritance   | Reuse code; create class hierarchies                      |
| Polymorphism  | Same interface, different behaviors                       |

#### What are Class and Object?

Class: A blueprint/template that defines attributes (data) and methods (behaviors).

Object: An instance of a class—actual entity created from the blueprint


```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def bark(self):
        return f"{self.name} says woof!"

dog1 = Dog("Buddy", 3)  # object
print(dog1.name)        # Buddy
print(dog1.bark())      # Buddy says woof!
```

#### What is __init__ (Constructor)?

__init__ is a special method called a constructor. It automatically runs when an object is created and initializes the object's attributes

```python
class Person:
    def __init__(self, name):
        self.name = name  # initializes the object's name
```

#### What is Encapsulation?

Encapsulation bundles data and methods within a class and hides internal details from outside access. It protects object state and prevents unexpected modification.

```python
class BankAccount:
    def __init__(self):
        self.__balance = 0  # private attribute (name mangling)
    
    def deposit(self, amount):
        self.__balance += amount
    
    def get_balance(self):
        return self.__balance

        
#Single underscore _balance: "internal use" by convention.
#Double underscore __balance: makes the attribute more private
```

#### What is Inheritance?

Inheritance allows a child class to inherit attributes and methods from a parent class, enabling code reuse and hierarchical relationships.

Types of inheritance in Python:

    Single

    Multiple

    Multilevel

    Hierarchical

    Hybrid.

```python
class Animal:
    def __init__(self, name):
        self.name = name
    
    def speak(self):
        return "Some sound"

class Dog(Animal):  # Dog inherits from Animal
    def speak(self):
        return "Woof!"

d = Dog("Buddy")
print(d.name)   # Buddy (from parent)
print(d.speak())  # Woof! (overridden)
```

#### What is Polymorphism?

Polymorphism allows objects of different classes to be treated as objects of a common class, with methods behaving differently based on the object type. It supports method overriding and method overloading (via default arguments).

```python
class Cat:
    def speak(self):
        return "Meow"

class Dog:
    def speak(self):
        return "Woof"

def make_animal_speak(animal):
    print(animal.speak())

make_animal_speak(Cat())  # Meow
make_animal_speak(Dog())  # Woof
```

#### What is Abstraction?

Abstraction hides complex implementation details and exposes only essential functionality. In Python, you use abstract base classes (ABC) to enforce interfaces.

```python
from abc import ABC, abstractmethod

class Shape(ABC):
    @abstractmethod
    def area(self):
        pass

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return 3.14 * self.radius ** 2

# Shape is abstract; you cannot instantiate it.
# Subclasses must implement area().
```

#### What Is self in Python?

self refers to the current instance of the class. It's the first parameter in instance methods and is used to access attributes and other methods of the object.



```python
class Person:
    def __init__(self, name):
        self.name = name  # self.name is an instance attribute

    def greet(self):
        return f"Hello, I'm {self.name}"
```
