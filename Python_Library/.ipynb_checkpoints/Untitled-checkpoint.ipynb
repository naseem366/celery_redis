```python
from pyspark.sql import SparkSession

# Create Spark session
spark = SparkSession.builder.appName("TestExample").getOrCreate()
```

```python
# Create DataFrame with 20 Records

data = [
    (1, "Alice", 25, "HR", 40000),
    (2, "Bob", 30, "IT", 60000),
    (3, "Charlie", 28, "Finance", 50000),
    (4, "David", 35, "IT", 70000),
    (5, "Eva", 22, "HR", 35000),
    (6, "Frank", 40, "Finance", 80000),
    (7, "Grace", 29, "IT", 62000),
    (8, "Helen", 31, "HR", 45000),
    (9, "Ian", 27, "Finance", 52000),
    (10, "Jack", 33, "IT", 68000),
    (11, "Kathy", 26, "HR", 39000),
    (12, "Leo", 38, "Finance", 75000),
    (13, "Mona", 24, "IT", 58000),
    (14, "Nina", 32, "HR", 42000),
    (15, "Oscar", 36, "Finance", 72000),
    (16, "Paul", 23, "IT", 55000),
    (17, "Queen", 34, "HR", 46000),
    (18, "Rick", 29, "Finance", 51000),
    (19, "Sara", 28, "IT", 63000),
    (20, "Tom", 37, "HR", 48000)
]

columns = ["id","name","age","department","salary"]
df = spark.createDataFrame(data,columns)
```

```python
emp_df = df
```

```python
# Department DataFrame
departments = [
    ("HR", "New York"),
    ("IT", "San Francisco"),
    ("Finance", "Chicago"),
    ("Marketing", "Los Angeles")
]
dept_columns = ["department", "location"]
dept_df = spark.createDataFrame(departments, dept_columns)
```

```python
emp_df.show()
```

```python
dept_df.show()
```

```python
inner_join = emp_df.join(dept_df, on= "department",how = "inner")
inner_join.show()
```

```python
left_join = emp_df.join(dept_df, on = "department", how = "left")
left_join.show()
```

```python
right_join = emp_df.join(dept_df, on = "department", how = "right")
right_join.show()
```

```python
full_join_df = emp_df.join(dept_df, on="department", how="outer")
full_join_df.show()
```

```python
# Add location to employee
emp_df2 = emp_df.withColumn("location", when(col("department")=="HR","New York")
                                        .when(col("department")=="IT","San Francisco")
                                        .otherwise(None))

multi_join_df = emp_df2.join(dept_df, on=["department","location"], how="inner")
multi_join_df.show()
```
