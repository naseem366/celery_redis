```python
df = pd.DataFrame(data,columns='A B C'.split())
```

```python
df
```

```python
# Create the DataFrame
data = {
    'A':[1,2,np.nan],
    'B':[5,np.nan,np.nan],
    'C':[1,2,3]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(axis=1)
```

### Group By

```python
import pandas as pd
# Create dataframe
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Sam','Charlie','Amy','Vanessa','Carl','Sarah'],
       'Sales':[200,120,340,124,243,350]}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.groupby('Company')
```

```python
by_compy = df.groupby('Company')
```

```python
mean_std = df.groupby('Company')['Sales']
```

```python
mean_std.sum()
```

```python
mean_std.std()
```

```python
mean_std.mean()
```

```python
by_compy.min()
```

```python
by_compy.max()
```

```python
by_compy.count()
```

```python
by_compy.describe()
```

```python
by_compy.describe().transpose()
```

```python
by_compy.describe().transpose()['GOOG']
```
